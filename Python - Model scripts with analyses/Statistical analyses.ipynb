```python
#Aminosyrer
#using the scaled values of amino acid concentrations I am performing a one-way T-test to see if the results are statistically significant from the original model. 
#I am also performing an anova analysis between media compositions. 

import pandas as pd
from scipy import stats

# Read the Excel spreadsheet into a pandas DataFrame
df = pd.read_excel("C:/Users/sofie/OneDrive/Documents/Master - Data/Aminoacid_concentrations.xlsx", "Glucose")

Expected_AA = [0.5051, 0.3653, 0.2801, 0.2801, 0.1073, 0.4928, 0.4928, 0.7723, 0.1546, 0.5107, 0.6555, 0.6114, 0.2145, 0.3329, 0.3041, 0.4091, 0.3526, 0.1028, 0.2097, 0.5807]

# Create an empty dictionary
concentration_dict = {}


# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Extract the amino acid name from the first column
    amino_acid = row['AA']
    
    # Extract the concentrations from the respective columns
    concentrations = [row['concentration 1'], row['concentration 2'], row['concentration 3']]
    
    # Add the amino acid and concentrations as key-value pairs in the dictionary
    concentration_dict[amino_acid] = concentrations

for expected_AA, (coefficient, values) in zip(Expected_AA, concentration_dict.items()):
    t_statistic, p_value = stats.ttest_1samp(values, expected_AA)
    
    print("Coefficient:", coefficient)
    print("Expected Value:", expected_AA)
    print("T-statistic:", t_statistic)
    print("P-value:", p_value)
    print()
```
